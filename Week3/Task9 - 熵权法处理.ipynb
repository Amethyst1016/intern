```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('df_final.csv')
df = df.iloc[:, 1:]
df.index = df['class0']
df = df[df.columns[1:]]
df
```

### 归一化

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()
df[df.columns] = sc.fit_transform(df[df.columns])
df
```

### 计算比重

```python
df_prop = df.apply(lambda x: x / x.sum(), axis=0)
df_prop = df_prop.replace(np.NaN, 0)
df_prop
```

### 计算每个指标的熵值

```python
K = 1/np.log(len(df_prop))
tmp = df_prop*np.log(df_prop)
tmp = tmp.replace(np.NaN, 0)
entropy = -K*(tmp.sum(axis=0))
entropy
```

### 计算每个指标的权重

```python
weight = (1 - entropy) / np.sum(1 - entropy)
weight
```

### 指标加权计算score

```python
score = df_prop.apply(lambda x: np.sum(100 * x * weight), axis=1)
score.nlargest(5)
```
