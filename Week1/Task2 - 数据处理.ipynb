```python
import pandas as pd
import numpy as np

user = pd.read_csv('user.csv')
user = user.iloc[:, 1:]
user.head()
```

```python
ad = pd.read_csv('ad.csv')
ad.head()
```

```python
click = pd.read_csv('click.csv')
click = click.iloc[:, 1:]
cols = np.array(click.columns)
cols[0] = 'userid'
click.columns = cols
click.head()
```

```python
# 合并三张表
user_ad = pd.merge(pd.merge(user, click, on='userid'), ad, on='ad_id')
display(user_ad.head())
len(user_ad)
```

```python
# 删除性别和年龄层级为空值的记录
user_ad.dropna(subset=['性别','年龄层级'], inplace = True)
display(user_ad.head())
len(user_ad) # 性别和年龄层级字段没有空值
```

```python
# 删除异常值（price<=100）
user_ad = user_ad.loc[~(user_ad['price']<=100)]
len(user_ad)
```

```python
# 对price字段进行分区处理
user_ad['price_range'] = pd.cut(x = user_ad['price'], bins = [100,200,300,800,1600,3200,6400,12800,max(user_ad['price'])])
user_ad.head()
```

```python
user_ad['price_labels'] = pd.Categorical(user_ad['price_range']).codes
user_ad.head()
```

```python
user_ad.to_csv('user_ad.csv')
```
